```python
# Renseigner les valeurs manquantes 

cat_data = []

num_data = []

for i, j in enumerate(df.dtypes):
    if j == object: 
        cat_data.append(df.iloc[:, i])
    else:
        num_data.append(df.iloc[:,i])

cat_data = pd.DataFrame(cat_data).transpose()
num_data = pd.DataFrame(num_data).transpose()
```

```python
# Pour les variables categoriques, on va remplacer les valeurs manquantes par celles des valeurs qui se repètent

cat_data = cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data.head()
```

```python
# Pour les variables numériques
# on remplace la valeur manquante par la valeur précédente 
# backward fill" (remplissage arrière).

num_data.fillna(method='bfill', inplace=True)
num_data.info()
```

```python
 # discréditer les variables 

target_value ={'Y':1, 'N':0}

target = cat_data['Loan_Status']

#cat_data.drop('Loan_Status', axis=1, inplace=True)
target = target.map(target_value)
target
```

```python
# je remplace les valeurs catégoriques par des valeurs numériques

le = LabelEncoder()

for i in cat_data:
    cat_data[i] = le.fit_transform(cat_data[i])

cat_data.head()
```

```python
# je supprime les variables inutiles

cat_data.drop('Loan_ID', axis=1, inplace=True)
```

```python
# je concatène 

X = pd.concat([cat_data, num_data], axis=1)

X.head()
```
